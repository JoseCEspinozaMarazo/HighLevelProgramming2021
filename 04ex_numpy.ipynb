5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
a = np.arange(1,11)
b = np.arange(1,11)
result_matrix=a[:, np.newaxis] * b

print(result_matrix)

#Finding the trace
print("\nThe trace of the matrix is:",result_matrix.trace())

#the anti-diagonal is result_matrix[0,-1],[1,-2]...[9,-10], this becomes [i][-(i+1)]
anti_diagonal=[result_matrix[i][-(i+1)] for i in range(len(result_matrix))]
print("\nThe anti diagonal of the matrix is:\n",anti_diagonal)

#for the diagonal plus 1 this becomes[1,0],[2,1]...[9,8], this becomes [i+1][i]
#since there are only 9 elements in this case the "in range" becomes len(result_matrix)-1
diagonal_plus=[result_matrix[i+1][i] for i in range(len(result_matrix)-1)]
print("\nThe diagonal plus 1 of the matrix is:\n",diagonal_plus)
````

6\. Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
#Building the 2d grid
import pandas

#Using the abreviations for the cities, Chicago=chi,los Angeles LA and so on
city_names=["CHI", "SPFLD", "STL", "TU", "Ok"
            , "AMA", "SF", "ABQ", "FLG", "LA"]

citiesA = np.array([0, 198, 303, 736,871,1175,1475,1544,1913,2448])
citiesB = np.array([0, 198, 303, 736,871,1175,1475,1544,1913,2448])
#using broadcast to get the 2d grid
broadcasting=abs(citiesA[:, np.newaxis]- citiesB)
#Using panda to get labels for row and columns
result = pandas.DataFrame(broadcasting, columns=city_names, index=city_names)
print("The distance between cities in miles are:")
display(result)


#1 mile is 1.60934 KM. rounding it to 1.6
KMresults=result*1.6

print("\n The distance between cities in Kilometers are:")
display(KMresults)
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
from timeit import default_timer as timer
import math

def sieve1(N):
    alist=[]
    
    eratos=np.arange(2,N+1)
    for i in eratos:
        if(i !=0):
            mask = (eratos % i == 0)
            alist.append(i)

        eratos[mask]=0
    primes=np.array(alist)
    return primes

N=99
begin= timer()
prime_num=sieve1(N)
ending=timer()
result= ending-begin
print("For an N of", N, "it takes:",result,"seconds")
print("\nWith a Higher N the function has a longer list making it take longer to obtain the results",
      "the part it take longest is with the number 2",
      "since the mask has to turn to false all values that are are multiples of 2\n")
print(prime_num)

#############################
#Optimized sieve

def imp_sieve(number):
    #number -1 since there 1 is not a prime number, start counting from 2
    A=np.ones((1, number-1), dtype=bool)
    result=np.arange(2,number+1)#From 2 to 99
    r_square=round(math.sqrt(number))
    for i in range(2,r_square+1):#from 2 to square of n, in case of N=99,10
        if A[0,i-2] ==True:#-2 since the start of the for is 2
            for j in range(2*i,number+1,i):
                A[0,j-2] =False# -2 same reason as the i
    mask=(A==True)#using mask to print only the prime numbers
    flat=mask.flatten()
    return result[flat]


begin= timer()
imp_result=imp_sieve(99)
ending=timer()
result= ending-begin
print(imp_result)
print("\nFor the improved sieve it took",result,"seconds")
print("The improved sieve took less time to execute")
````
